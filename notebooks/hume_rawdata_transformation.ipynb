```python
for filename in os.listdir(directory):
    input = os.path.join(directory, filename)
    output = os.path.join(directory, os.path.splitext(filename)[0]+ ".csv")

    # checking if it is a file
    if os.path.isfile(input):
        hume_prediction_ = open(str(input))
        hume_prediction_ = json.load(hume_prediction_)

        #hume_prediction_ = hume_prediction_['predictions']['facial']

        hume_prediction_ = hume_prediction_[0]['results']['predictions'][0]['models']['face']['grouped_predictions']

        df = pd.DataFrame.from_dict(hume_prediction_)

        predictions = pd.DataFrame()

        x = 0

        while x < len(df):
            placeholder_dataframe = pd.DataFrame.from_dict(df['predictions'][x])

            predictions = pd.concat([predictions, placeholder_dataframe], ignore_index= True)

            x = x+1



        for index, row in predictions.iterrows():
            for j in row['emotions']:
                emotion = j['name']
                score = j['score']
                predictions.loc[index, emotion] = score


        predictions = predictions.drop(columns = ['prob', 'box', 'emotions', 'facs', 'descriptions'], axis= 1)


        predictions.to_csv(output)



        
```

```python
combined_df = pd.DataFrame()

for filename in os.listdir(directory):
    if filename.endswith(".csv"):

        input_file = os.path.join(directory, filename)
        output = os.path.join(directory, os.path.splitext(filename)[0]+ "complete_video.csv")
        
        df = pd.read_csv(str(input_file))

        combined_df = pd.concat([combined_df, df], ignore_index= True)



combined_df.to_csv(output)
```
