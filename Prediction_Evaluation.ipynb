```python
import pandas as pd
import json
```

```python
example_prediction = open('data/c1_000.json')
```

```python
example_prediction = json.load(example_prediction)
```

```python
test = example_prediction[0]['results']['predictions'][0]['models']['face']['grouped_predictions'][0]

test['predictions'][0]['emotions']
```

```python
id_value = test['id']
predictions = test['predictions']

emotions = predictions[0]['emotions']

df = pd.DataFrame()

x = 0
while x < 48:
    for i, emotion in enumerate(emotions):
        name = emotions[x]['name']
        print(name)
        score = emotions[x]['score']
        print(score)


        df[f'emotion_{i+1}_name'] = name
        df[f'emotion_{i+1}_score'] = score

        x = x +1


df['id'] = id_value

print(df)
```
